### Measure of dispersion

How much your data is spread

Various Tech-

1) Variance----------  Measure the spread
2) Std deviation-----  It helps us to find out how much data is away from the mean

(meu- alt+230, sigma- alt+228, 2power- alt+0178)

Mathematical formula: 
    
    
    mean(µ)= Σx / N       --------This mean is known as population mean
    
    
    mean(x-bar)= Σx/n     --------This mean is known as sample mean
    
    
**Variance**:
    Population Variance: 
        σ²= Σ(x- µ)²/N
    
    
    
    
    Sample Variance: 
        v= (Σ(x- µ)²) / (n-1)
        
    
**Standard Deviation**:
    Population--------   σ= √ [Σ(x- µ)²/N]
    
    
    
    Sample-----          v=  √ [Σ(x- µ)²/n-1]

### Multiplicative probablity=>
Before this first we need to understand Independent and dependent event.

Independent Event: 
    
    Exp-   Roll a dice{1,2,3,4,5,6}
    
    P(1)= 1/6            P(2)=1/6               p(3)=1/6
    P(4)= 1/6            P(5)=1/6               p(6)=1/6
    
    So here occurance of first event does not impact the result of second event
    
    
    
    P(1 and 6)= P(1) * P(6)
              = 1/6 * 1/6
              = 1/36
    
    
    
Dependent Event:
    Exp-   Bag of Marbel{Red, White, Green, Black, Orange}
    
    P(Red)= 1/5         P(White)=1/4------------------
    
    So as we remove the red marbel so we have only 4 marbel so occurance of second event will be depend on first event.
    
    P(White and Red)= P(White) * P(Red/White)
                    = 1/5      *   1/4
                    = 1/20
            
This is known as **conditional probablity**            
This will also use in Naive Bayes
