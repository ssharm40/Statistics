First Qus--- **What is percentile?**

Exp- Data={2,2,3,4,5,5,6,7,8,8,8,9,9,10}
What is the percentile of 7?

Percentile means => How much data is less than the X value

percentile of value x = [#(num.) of value less than x    * 100]/n




percentile of value 7= (7*100)/14   => 50 percentile

so it's percentage data is less than 7




Now Qua is what is the 25th percentile of data:

Position= (percentile* n+1)/100

        => (25 * 15)/100 =>3.75(then consider it as 4, and get the 4 position from data that is 4)
        
        


Why we use this?  => To Remove the Outliers


For removing the outlier we need to identify the boundaries-

[Lower Boundaries--------------------------Upper Boundaries]



Formula for finding the fence

Lower fence=   Q1  - 1.5(IQR)

Upper Fence=   Q3  + 1.5(IQR)



Q1= 25 percentile
Q3= 75 percentile

IQR= Q3-Q1


All the variable that come outside of this range will be considered as Outlier






**Example** 
x={5,6,7,2,15,35,88,77,65,52,43,15,105,208,199,1000}

x={2,5,6,7,15,15,35,43,52,65,77,88,105,199,208,1000}

### Co-Variance:

It's use to quantify the relationship b/w two features

means If X value increase the Y value> ?


Exp------- Height(increase)===> Weight(increase)         Means Positive Co-Variance
            
           Marketing(increase) ====> Sales(Increase)     Means Positive Co-Variance
        
           Drinking (Increase) ====> Age(Decrease)       Means Negative Co-Variance
            
            
So finding the relationship b/w two feature is known as Co-Varaince



Cov(x,y)= [Σ (x-x_bar) * (y- y_bar)]/n-1

## Central Limit Theorem

Central limit theorem says that a variable X may or may not belong to gaussian distribution with (some mean and std), but if
we take multiple sample and size at least n>= 0 and take mean of that samples and start plotting them then it will
definitely follow the Gaussian Distribution with (mean or std)

Formula:
    
                                X ≈ GD(mean, std)
