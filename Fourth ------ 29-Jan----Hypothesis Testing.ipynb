## Hypothesis Testing

**Steps**

1) Define Null Hypotehsis (H0) 
2) Alternate Hypothesis (H1)
3) Statistical Experiment and get the p value(z-test, t-test, f-test, chi-square-test, anova test........)
4) Conclusion (Accept/ Reject the Null hypothesis)

## P- value

After performing statistical experiment(Z-Test, T-test) you will get some value, this value is known as P value

Suppose α= .05
then CI  = .95
then half -ve tail= .025
and  half +ve tail= .025
